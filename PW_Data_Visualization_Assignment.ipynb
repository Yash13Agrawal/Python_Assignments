**5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.**

````python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate a synthetic dataset with correlated features
# We can create a 2D array where columns are features
np.random.seed(0) # for reproducibility
data = np.random.rand(100, 5) # 100 samples, 5 features

# Introduce some correlation
data[:, 1] = data[:, 0] + np.random.normal(0, 0.1, 100)
data[:, 2] = -data[:, 0] + np.random.normal(0, 0.1, 100)

# Calculate the correlation matrix
corr_matrix = np.corrcoef(data.T)

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Synthetic Dataset')
plt.show()
````

**3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and
days.**


```
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)

```



````python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# To create a heatmap of sales across months and days, we need to aggregate the sales data
# Group by Month and Day and sum the Sales
heatmap_data = df.groupby(['Month', 'Day'])['Sales'].sum().reset_index()

# Create a pivot table for the heatmap
heatmap_pivot = heatmap_data.pivot(index='Month', columns='Day', values='Sales').fillna(0)

# Define the order of months
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
heatmap_pivot = heatmap_pivot.reindex(index=month_order)

# Create the heatmap using plotly.graph_objects
fig = go.Figure(data=go.Heatmap(
                   z=heatmap_pivot.values,
                   x=heatmap_pivot.columns,
                   y=heatmap_pivot.index,
                   colorscale='Viridis')) # You can choose a different colorscale

fig.update_layout(title='Sales Variation Across Months and Days',
                  xaxis_title='Day',
                  yaxis_title='Month')

fig.show()
````

**1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.**

````python
from bokeh.plotting import figure, show
import numpy as np

# Prepare the data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
````

**2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.**

````python
from bokeh.plotting import figure, show
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)
num_points = 100
data = {
    'x': np.random.rand(num_points) * 10,
    'y': np.random.rand(num_points) * 10,
    'sizes': np.random.rand(num_points) * 30,
    'colors': np.random.choice(['navy', 'firebrick', 'olive', 'teal'], num_points)
}
df = pd.DataFrame(data)

# Create a new plot
p = figure(title="Bokeh Scatter Plot with Varying Size and Color",
           x_axis_label='X', y_axis_label='Y')

# Add a scatter renderer with size and color mapped from columns
p.scatter(x='x', y='y', size='sizes', color='colors', source=df, legend_field="colors")

# Show the results
show(p)
````

**3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**


```
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```



````python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource from the data
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count')

# Add a vertical bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source)

# Show the results
show(p)
````

````python
from bokeh.plotting import figure, show
import numpy as np

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Bokeh Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad renderer for the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color="white")

# Show the results
show(p)
````

**5. Create a Bokeh heatmap using the provided dataset.**


```
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```



````python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis9

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for ColumnDataSource
source_data = {'x': [], 'y': [], 'value': []}
for i in range(data_heatmap.shape[0]):
    for j in range(data_heatmap.shape[1]):
        source_data['x'].append(xx[i, j])
        source_data['y'].append(yy[i, j])
        source_data['value'].append(data_heatmap[i, j])

source = ColumnDataSource(data=source_data)

# Create a new plot
p = figure(title="Bokeh Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1))

# Create the heatmap using rect glyph
mapper = linear_cmap(field_name='value', palette=Viridis9, low=min(source_data['value']), high=max(source_data['value']))

p.rect(x='x', y='y', width=x[1]-x[0], height=y[1]-y[0],
       source=source, fill_color=mapper, line_color=None)

# Show the results
show(p)
````
