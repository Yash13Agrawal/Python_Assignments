# **Question 1.** Discuss String slicinng and Provide Examples.

**Answer:** String slicing is a powerful feature in Python that allows you to extract a portion of a string by specifying a range of indices. The syntax for slicing is:

`**string[start:end:step]**`

*   start: The index where the slice begins (inclusive).
*   end: The index where the slice ends (exclusive).
# *   step: The interval between each index in the slice (optional).

````python
#Example:
text = "Hello, World!"
slice1 = text[0:5]  # 'Hello'
slice2 = text[7:12]  # 'World'
slice3 = text[:]  # 'Hello, World!'
slice4 = text[-6:-1]  # 'World'
slice5 = text[-6:]  # 'World!'
slice6 = text[::2]  # 'Hlo ol!'
slice7 = text[1:10:2]  # 'el,W '
````

# **Question 2.** Explain the key features of list in python.

**Answer:**
Lists in Python are versatile and powerful data structures that allow you to store collections of items. Here are some key features of lists:

  1. **Ordered Collection**
- Lists maintain the order of items, meaning that the elements can be accessed based on their index, which is zero-based.

  2. **Mutable**
- Lists are mutable, allowing you to modify their contents after creation. You can add, remove, or change elements.

  3. **Dynamic Sizing**
- Lists can grow or shrink in size. You can add elements without needing to define a fixed size beforehand.

  4. **Heterogeneous Elements**
- Lists can contain elements of different data types, including integers, strings, floats, or even other lists.

  5. **Indexing and Slicing**
- You can access elements using their index, and slicing allows you to retrieve a portion of the list. For example, `my_list[1:3]` gives you a sublist.

  6. **Built-in Methods**
- Python lists come with a variety of built-in methods for common operations, such as:
  - `append()`: Adds an element to the end.
  - `insert()`: Inserts an element at a specified position.
  - `remove()`: Removes the first occurrence of a value.
  - `pop()`: Removes and returns an element at a given index.
  - `sort()`: Sorts the list in ascending order.

  7. **List Comprehensions**
- You can create lists using concise syntax known as list comprehensions, which allow you to generate lists dynamically. For example:
  ```python
  squares = [x**2 for x in range(10)]  # Generates [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
  ```

  8. **Nested Lists**
- Lists can contain other lists as elements, allowing you to create multi-dimensional data structures.



````python
# Here’s a quick example demonstrating some of these features:

# ```python
# Creating a list
my_list = [1, 'hello', 3.14, [2, 3], True]

# Accessing elements
print(my_list[1])          # Output: 'hello'
print(my_list[3][0])      # Output: 2

# Modifying the list
my_list.append('new item')  # Adding an item
print(my_list)              # Output: [1, 'hello', 3.14, [2, 3], True, 'new item']

# Slicing the list
sub_list = my_list[1:4]    # Output: ['hello', 3.14, [2, 3]]

# Using list comprehension
squares = [x**2 for x in range(5)]  # Output: [0, 1, 4, 9, 16]
````

# **Question 3.** Describe how to access, modify and delete elements in a list with example.

**Answer:**
Accessing, modifying, and deleting elements in a list are fundamental operations in Python. Here’s a detailed explanation with examples for each operation.

### Accessing Elements

You can access elements in a list using indexing. Python uses zero-based indexing, so the first element is at index 0.

#### Example:
```python
my_list = [10, 20, 30, 40, 50]

# Accessing elements
first_element = my_list[0]  # 10
third_element = my_list[2]   # 30
last_element = my_list[-1]    # 50 (negative indexing)

print(first_element)  # Output: 10
print(third_element)  # Output: 30
print(last_element)   # Output: 50
```

### Modifying Elements

You can modify elements by assigning a new value to a specific index.

#### Example:
```python
my_list = [10, 20, 30, 40, 50]

# Modifying elements
my_list[1] = 25  # Change the second element
my_list[3] = 45  # Change the fourth element

print(my_list)  # Output: [10, 25, 30, 45, 50]
```

### Deleting Elements

You can delete elements from a list using several methods, such as `del`, `remove()`, or `pop()`.

1. **Using `del`**: Removes an element at a specific index.
2. **Using `remove()`**: Removes the first occurrence of a value.
3. **Using `pop()`**: Removes and returns the element at a specific index (defaults to the last element if no index is specified).

#### Examples:

**Using `del`:**
```python
my_list = [10, 20, 30, 40, 50]

# Deleting an element
del my_list[2]  # Removes the third element

print(my_list)  # Output: [10, 20, 40, 50]
```

**Using `remove()`:**
```python
my_list = [10, 20, 30, 40, 50]

# Removing a specific value
my_list.remove(30)  # Removes the first occurrence of 30

print(my_list)  # Output: [10, 20, 40, 50]
```

**Using `pop()`:**
```python
my_list = [10, 20, 30, 40, 50]

# Popping the last element
popped_element = my_list.pop()  # Removes and returns 50
print(popped_element)  # Output: 50
print(my_list)  # Output: [10, 20, 30, 40]

# Popping an element at a specific index
popped_element_index = my_list.pop(1)  # Removes and returns the second element (20)
print(popped_element_index)  # Output: 20
print(my_list)  # Output: [10, 30, 40]
```

### Summary

- **Accessing**: Use indexing to get values from a list.
- **Modifying**: Assign a new value to a specific index to change an element.
- **Deleting**: Use `del`, `remove()`, or `pop()` to remove elements from a list.

These operations make lists flexible and powerful for managing collections of data in Python.

# **Question 4.** Compare and contrast tuples and lists with examples.
# **Answer:**  
Tuples and lists are both built-in data structures in Python that can store collections of items. However, they have some key differences in terms of mutability, syntax, and use cases. Here’s a detailed comparison:

### 1. **Mutability**

- **Lists**: Mutable, meaning you can modify their contents (add, remove, or change elements).
- **Tuples**: Immutable, meaning once they are created, their contents cannot be changed.

**Example:**
```python
# List
my_list = [1, 2, 3]
my_list[0] = 10  # Modifying an element
my_list.append(4)  # Adding an element
print(my_list)  # Output: [10, 2, 3, 4]

# Tuple
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This would raise a TypeError
```

### 2. **Syntax**

- **Lists**: Defined using square brackets `[]`.
- **Tuples**: Defined using parentheses `()`.

**Example:**
```python
# List
my_list = [1, 2, 3]

# Tuple
my_tuple = (1, 2, 3)
```

### 3. **Performance**

- **Lists**: Slightly slower than tuples because of their mutability.
- **Tuples**: Generally faster and more memory-efficient due to their immutability.

### 4. **Use Cases**

- **Lists**: Suitable for collections of items that may need to change (e.g., a list of user inputs).
- **Tuples**: Better for fixed collections of items that should not change (e.g., coordinates, RGB color values).

**Example:**
```python
# List example
user_inputs = [input("Enter name: "), input("Enter age: ")]
print(user_inputs)

# Tuple example
coordinates = (10.0, 20.0)
print(coordinates)
```

### 5. **Methods**

- **Lists**: Have a variety of methods like `append()`, `remove()`, and `sort()`.
- **Tuples**: Have fewer methods, mainly `count()` and `index()`.

**Example:**
```python
# List methods
my_list = [3, 1, 2]
my_list.sort()  # Sorts the list
print(my_list)  # Output: [1, 2, 3]

# Tuple methods
my_tuple = (1, 2, 2, 3)
count_of_twos = my_tuple.count(2)  # Counts occurrences of 2
print(count_of_twos)  # Output: 2
```



# **Question 5.** Describe the key features of sets and provide examples of their use.
# **Answer:**
Sets are a built-in data structure in Python that store unordered collections of unique items. Here are the key features of sets, along with examples of their use:

# Key Features of Sets

**Unordered:**
Sets do not maintain any specific order for their elements. The items cannot be accessed by index.

**Unique Elements:**
Sets automatically enforce uniqueness; duplicate items are not allowed. If you try to add a duplicate, it will simply be ignored.  

**Mutable:**
Sets are mutable, meaning you can add or remove elements after creation.

**Dynamic Sizing:**
Sets can grow and shrink in size as elements are added or removed.

**Supports Set Operations:**
Sets support mathematical operations such as union, intersection, difference, and symmetric difference, which can be very useful in various scenarios.

**No Indexing or Slicing:**
Since sets are unordered, you cannot access elements by index or slice them.

**Hashable:**
The elements of a set must be hashable, which means they must be immutable (like strings, numbers, and tuples).


````python
#Example:
my_set = {1, 2, 3, 4, 5}
print(my_set)  # Output: {1, 2, 3, 4, 5}

#Adding and Removing Elements
my_set.add(6)          # Adding an element
my_set.add(2)          # Trying to add a duplicate (will be ignored)
my_set.remove(1)       # Removing an element
print(my_set)

#Union: Combines two sets, removing duplicates.
set_a = {1, 2, 3}
set_b = {3, 4, 5}
union_set = set_a.union(set_b)
print(union_set)  # Output: {1, 2, 3, 4, 5}

#Intersection: Returns elements common to both sets.
intersection_set = set_a.intersection(set_b)
print(intersection_set)  # Output: {3}

````

# **Question 7.** Describe how to add, modify and delete iyems in a dictionary with examples.
# **Answer:**

Dictionaries in Python are mutable data structures that store key-value pairs. You can easily add, modify, and delete items in a dictionary. Here’s how to do each of these operations, along with examples.

# Adding Items

You can add new key-value pairs to a dictionary by using the assignment operator.

**Example:**
```python
# Creating a dictionary
my_dict = {'name': 'Alice', 'age': 25}

# Adding a new item
my_dict['city'] = 'New York'

print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
```

# Modifying Items

To modify an existing item, you can simply assign a new value to the key.

**Example:**
```python
# Modifying an existing item
my_dict['age'] = 26  # Update the age

print(my_dict)  # Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}
```

### Deleting Items

You can delete items from a dictionary using several methods: `del`, `pop()`, or `popitem()`.

1. **Using `del`**: Removes the item with the specified key.
2. **Using `pop()`**: Removes and returns the value associated with the specified key.
3. **Using `popitem()`**: Removes and returns the last inserted key-value pair (available in Python 3.7+).

# **#### Example of Deleting Items:**

**Using `del`:**
```python
# Deleting an item using del
del my_dict['city']

print(my_dict)  # Output: {'name': 'Alice', 'age': 26}
```

**Using `pop()`:**
```python
# Deleting an item using pop
age = my_dict.pop('age')  # Removes 'age' and returns its value

print(age)         # Output: 26
print(my_dict)    # Output: {'name': 'Alice'}
```

**Using `popitem()`:**
```python
# Deleting the last inserted item using popitem
item = my_dict.popitem()  # Removes and returns the last item

print(item)       # Output: ('name', 'Alice')
print(my_dict)    # Output: {}
```




# **Question 8.** Discuss the importance of dictionary keys being immutable and provide examples.

# Answer:
In Python, dictionary keys must be immutable data types. This immutability is crucial for several reasons, primarily related to the functionality and integrity of dictionaries. Here’s a detailed discussion of the importance of immutable keys, along with examples.

### Importance of Immutable Keys

1. **Hashability**:
   - Dictionary keys are hashed to create a unique identifier for each key. This hash is used to quickly access values. If a key were mutable and changed, it would affect its hash value, leading to inconsistencies and making it impossible to locate the associated value.

   **Example**:
   ```python
   my_dict = { (1, 2): "Point A" }  # Tuple (1, 2) is immutable and can be a key
   # my_dict[(1, 2)]  # Valid lookup
   
   # my_dict[[1, 2]] = "Point B"  # This would raise a TypeError, since lists are mutable
   ```

2. **Data Integrity**:
   - Immutable keys ensure that once a key is set, it cannot be altered, which maintains the integrity of the dictionary. This prevents unexpected behavior caused by unintentional changes to keys.

   **Example**:
   ```python
   my_dict = {'name': 'Alice'}
   # my_dict['name'][0] = 'B'  # This would not work because strings are immutable, ensuring the key stays consistent
   ```

3. **Consistent Behavior**:
   - The requirement for immutable keys leads to predictable behavior in dictionaries. You can rely on the fact that once you insert a key-value pair, the key will always point to the same value unless explicitly changed.

   **Example**:
   ```python
   my_dict = {1: 'one', 2: 'two'}
   print(my_dict[1])  # Output: 'one'
   # my_dict[1] = 'uno'  # Changing the value associated with the immutable key 1
   ```
